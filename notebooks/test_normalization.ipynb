```python
from anyascii import anyascii

# checking for Hindi script
hindi_uni = anyascii('नमस्ते विद्यार्थी')

print("The translation from hindi Script : "
      + str(hindi_uni))

# checking for Punjabi script
pun_uni = anyascii('ਸਤਿ ਸ੍ਰੀ ਅਕਾਲ')

print("The translation from Punjabi Script : "
      + str(pun_uni))
```

```python
from anyascii import anyascii

# 1. Testing ions with superscripts (Na⁺, Cl⁻)
# Using actual Unicode superscript characters '⁺' and '⁻'
ions = anyascii("Na⁺ and Cl⁻")
print("1. Superscripts (Ions): " + ions)

# 2. Testing Greek letters (alpha, beta, gamma, omega)
greek = anyascii("α, β, γ, Ω")
print("2. Greek Letters:       " + greek)

# 3. Testing the Celsius symbol (°C) and degree sign (°)
celsius = anyascii("The temperature is 25°C (or 180°)")
print("3. Degree / Celsius:    " + celsius)

# 4. Testing French accents (é, è, à, ç, ù, ï, ô)
accents = anyascii("Café, crème, garçon, Noël, vôtre")
print("4. French Accents:      " + accents)

# 5. Testing 'weird' dashes and hyphens (en-dash, em-dash, figure dash)
# –, —, ‒ are different Unicode dash variants
dashes = anyascii("Dash types: – (en), — (em), ‒ (figure)")
print("5. Weird Dashes:        " + dashes)
```
